# 02 - YOLOv8 Fine-tuning

Fine-tune YOLOv8 for traffic-specific object detection.
Additional classes: traffic_cone, debris, fire, smoke.

```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Load pretrained YOLOv8
model = YOLO('yolov8s.pt')
print(f'Model loaded: {model.model_name}')
```

```python
# Test on a sample traffic image
results = model('https://images.wsdot.wa.gov/nw/005vc16757.jpg')
for r in results:
    print(f'Detected {len(r.boxes)} objects')
    for box in r.boxes:
        cls = int(box.cls[0])
        conf = float(box.conf[0])
        name = model.names[cls]
        print(f'  {name}: {conf:.2%}')
```
