# 01 - Data Exploration

Explore the crash detection datasets:
- CADP (Car Accident Detection and Prediction)
- DoTA (Detection of Traffic Anomaly)
- Normal traffic captures from DOT cameras

```python
import os
from pathlib import Path
import cv2
import numpy as np
import matplotlib.pyplot as plt

DATA_DIR = Path('../data')
RAW_DIR = DATA_DIR / 'raw'
PROCESSED_DIR = DATA_DIR / 'processed'

print(f'Raw data dir exists: {RAW_DIR.exists()}')
print(f'Processed data dir exists: {PROCESSED_DIR.exists()}')
```

```python
# Count available images per class
for label in ['crash', 'normal']:
    for split in ['train', 'val', 'test']:
        path = PROCESSED_DIR / 'splits' / label / split
        if path.exists():
            count = len(list(path.glob('*.jpg')))
            print(f'{label}/{split}: {count} images')
        else:
            print(f'{label}/{split}: not found')
```

```python
# Display sample images
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
fig.suptitle('Sample Images', fontsize=16)

for i, label in enumerate(['crash', 'normal']):
    img_dir = PROCESSED_DIR / 'classifier' / label
    if not img_dir.exists():
        continue
    images = list(img_dir.glob('*.jpg'))[:4]
    for j, img_path in enumerate(images):
        img = cv2.imread(str(img_path))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        axes[i, j].imshow(img)
        axes[i, j].set_title(f'{label}')
        axes[i, j].axis('off')

plt.tight_layout()
plt.show()
```
